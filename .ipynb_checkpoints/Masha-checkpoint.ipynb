```python
from pathlib import Path
import torch
from PIL import Image
from tqdm import tqdm
from models.experimental import attempt_load
from utils.general import non_max_suppression
from utils.torch_utils import select_device

# Путь к изображениям
img_folder = 'путь_к_папке_с_изображениями'

# Загрузка предобученной модели YOLOv5
device = select_device('')
model = attempt_load('yolov5s.pt', map_location=device)

# Режим инференса
model.eval()

# Путь для записи результатов
output_file = 'output.txt'

# Проход по изображениям в папке
with open(output_file, 'w') as f:
    for img_path in tqdm(Path(img_folder).glob('*.jpg')):
        img = Image.open(img_path)
        pred = model(img, size=640)  # Размер изображения для обработки
        pred = non_max_suppression(pred, 0.4, 0.5)[0]  # Убираем дублирующиеся боксы

        if pred is not None:
            for det in pred:
                x1, y1, x2, y2, conf, class_id = det.tolist()
                f.write(f'{img_path} {int(x1)},{int(y1)},{int(x2-x1)},{int(y2-y1)},{int(class_id)}\n')
```

```python
# Import required packages
import cv2
import pytesseract

# Read image from which text needs to be extracted
img = cv2.imread("dnr-train/train/IMG_20240408 (14).jpg")

# Preprocessing the image starts

# Convert the image to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Performing OTSU threshold
ret, thresh1 = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU | cv2.THRESH_BINARY_INV)

# Specify structure shape and kernel size. 
# Kernel size increases or decreases the area 
# of the rectangle to be detected.
# A smaller value like (10, 10) will detect 
# each word instead of a sentence.
rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (18, 18))

# Applying dilation on the threshold image
dilation = cv2.dilate(thresh1, rect_kernel, iterations = 1)

# Finding contours
contours, hierarchy = cv2.findContours(dilation, cv2.RETR_EXTERNAL, 
												cv2.CHAIN_APPROX_NONE)

# Creating a copy of image
im2 = img.copy()

# A text file is created and flushed
file = open("recognized.txt", "w+")
file.write("")
file.close()

# Looping through the identified contours
# Then rectangular part is cropped and passed on
# to pytesseract for extracting text from it
# Extracted text is then written into the text file
for cnt in contours:
	x, y, w, h = cv2.boundingRect(cnt)
	
	# Drawing a rectangle on copied image
	rect = cv2.rectangle(im2, (x, y), (x + w, y + h), (0, 255, 0), 2)
	
	# Cropping the text block for giving input to OCR
	cropped = im2[y:y + h, x:x + w]
	
	# Open the file in append mode
	file = open("recognized.txt", "a")
	
	# Apply OCR on the cropped image
	text = pytesseract.image_to_string(cropped, lang='rus')
	
	# Appending the text into file
	file.write(text)
	file.write("\n")
	
	# Close the file
	file.close
```

```python
import cv2
import random
from ultralytics import YOLO

def process_image_with_tracking(model, input_image_path):
    # Read the input image
    frame = cv2.imread(input_image_path)

    # Track objects in the image
    results = model(frame)

    if results.xyxy[0] is not None: # Check if there are any detections
        boxes = results.xyxy[0].cpu().numpy().astype(int)
        ids = results.track_id[0].cpu().numpy().astype(int)

        for box, id in zip(boxes, ids):
            # Generate a random color for each object based on its ID
            random.seed(int(id))
            color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

            cv2.rectangle(frame, (box[0], box[1]), (box[2], box[3]), color, 2)
            cv2.putText(
                frame,
                f"Id {id}",
                (box[0], box[1]),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.5,
                (0, 255, 255),
                2,
            )

    # Display the processed image
    cv2.imshow("Processed Image", frame)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage:
model = YOLO('runs/detect/train/weights/best.pt')
model.fuse()
process_image_with_tracking(model, "test.jpg")
```

```python
from ultralytics import YOLO
import cv2
import math

import nltk
from nltk.corpus import stopwords
from pymystem3 import Mystem
from string import punctuation
import pickle

import easyocr

reader = easyocr.Reader(["ru", "en"])
modelPriceTag = YOLO('models/detectPriceTag.pt')
modelPriceNameSocialPrice = YOLO('models/detectPriceNameSocialPrice.pt')
modelClassificationTitle = pickle.load(open("models/ClassificationTitle.pickle", 'rb'))
ClassificationTitleTfidfVectorizer = pickle.load(open("models/ClassificationTitleTfidfVectorizer.pickle", 'rb'))

nltk.download('stopwords')
mystem = Mystem()
russian_stopwords = stopwords.words("russian")


def preprocess_text(text):
    text = str(text)
    tokens = mystem.lemmatize(text.lower())
    tokens = [token for token in tokens if token not in russian_stopwords\
              and token != " " \
              and len(token)>=3 \
              and token.strip() not in punctuation \
              and token.isdigit()==False]
    text = " ".join(tokens)
    return text

def convert_to_int(string):
    if '.' not in string:
        digits = ''
        for char in string:
            if char.isdigit():  # Проверяем, является ли символ цифрой
                digits += char
        if digits:  # Если строка не пуста
            return int(digits)
        else:
            return None  # Можно вернуть None, если в строке нет цифр
    else:
        return None

def detectPriceTag(pathFile):
    model = YOLO('models/detectPriceTag.pt')
    model.fuse()

    frame = cv2.imread(pathFile)

    results = model.track(frame, iou=0.4, conf=0.5, persist=True, imgsz=608, verbose=False, tracker="models/botsort.yaml")
    praiceTags = list()

    if results[0].boxes.id != None :
        boxes = results[0].boxes.xyxy.cpu().numpy().astype(int)
        ids = results[0].boxes.id.cpu().numpy().astype(int)

        for box, idd in zip(boxes, ids):
            cropped = frame[box[1]:box[3], box[0]:box[2]]
            praiceTags.append(cropped)

    return praiceTags

def detectPriceNameSocialPrice(pathFile):
    model = YOLO('models/detectPriceNameSocialPrice.pt')
    model.fuse()

    priceTag = detectPriceTag(pathFile)
    frame = priceTag[math.floor((len(priceTag)/2))]

    results = model.track(frame, iou=0.4, conf=0.5, persist=True, imgsz=608, verbose=False, tracker="models/botsort.yaml")
    title = ""
    category = ""
    price = 0
    socialPrice = ""
    flag = False

    if results[0].boxes.id != None :
        boxes = results[0].boxes.xyxy.cpu().numpy().astype(int)
        ids = results[0].boxes.id.cpu().numpy().astype(int)

        for box, idx in zip(boxes, ids):
            cropped = frame[box[1]:box[3], box[0]:box[2]]

            if idx == 2:
                title += "".join(reader.readtext(cropped, detail=0, paragraph=True))
            elif idx == 1:
                flag_price = True
                readerr = reader.readtext(cropped, detail=0)
                for item in readerr:
                    if convert_to_int(item) != None and flag_price:
                        price = f'{convert_to_int(item)}'
                        flag_price = False
                    elif convert_to_int(item) != None and (not(flag_price)) :
                        price = price + "." + f'{convert_to_int(item)}'
            elif idx == 3:
                socialPrice += "".join(reader.readtext(cropped, detail=0, paragraph=True))


    new_category = [preprocess_text(title)]  # List of new titles
    new_category_tfidf = ClassificationTitleTfidfVectorizer.transform(new_category)
    predictions = modelClassificationTitle.predict(new_category_tfidf)


    if socialPrice is not None:
        flag == True
        
    return {
        "title": title,
        "category": predictions[0],
        "price": price,
        "socialPrice": flag,
    }

for i in range(1, 100):
    detectPriceNameSocialPrice("dataset/dataset-PriseTags/42.jpg")
# praiceTags = detectPriceNameSocialPrice('dataset/dataset-PriseTags/42.jpg')
# print(praiceTags)
```

```python
import cv2
import os
import csv
import math

import nltk
from nltk.corpus import stopwords
from pymystem3 import Mystem
from string import punctuation
import pickle
import pandas as pd

import easyocr
from ultralytics import YOLO
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

reader = easyocr.Reader(["ru", "en"])
modelPriceTag = YOLO('models/detectPriceTag.pt')
modelPriceNameSocialPrice = YOLO('models/detectPriceNameSocialPrice.pt')
modelClassificationTitle = pickle.load(open("models/ClassificationTitle.pickle", 'rb'))
ClassificationTitleTfidfVectorizer = pickle.load(open("models/ClassificationTitleTfidfVectorizer.pickle", 'rb'))

nltk.download('stopwords')
mystem = Mystem()
russian_stopwords = stopwords.words("russian")


def preprocess_text(text):
    text = str(text)
    tokens = mystem.lemmatize(text.lower())
    tokens = [token for token in tokens if token not in russian_stopwords \
              and token != " " \
              and len(token) >= 3 \
              and token.strip() not in punctuation \
              and token.isdigit() == False]
    text = " ".join(tokens)
    return text


def convert_to_int(string):
    if '.' not in string:
        digits = ''
        for char in string:
            if char.isdigit():  # Проверяем, является ли символ цифрой
                digits += char
        if digits:  # Если строка не пуста
            return int(digits)
        else:
            return None  # Можно вернуть None, если в строке нет цифр
    else:
        return None


def detectPriceTag(pathFile):
    model = YOLO('models/detectPriceTag.pt')
    model.fuse()

    frame = cv2.imread(pathFile)

    results = model.track(frame, iou=0.4, conf=0.5, persist=True, imgsz=608, verbose=False,
                          tracker="models/botsort.yaml")
    praiceTags = list()

    if results[0].boxes.id != None:
        boxes = results[0].boxes.xyxy.cpu().numpy().astype(int)
        ids = results[0].boxes.id.cpu().numpy().astype(int)

        for box, idd in zip(boxes, ids):
            cropped = frame[box[1]:box[3], box[0]:box[2]]
            praiceTags.append(cropped)

    return praiceTags


def detectPriceNameSocialPrice(pathFile):
    model = YOLO('models/detectPriceNameSocialPrice.pt')
    model.fuse()

    priceTag = detectPriceTag(pathFile)
    if len(priceTag) == 0:
        return {
        "title": "",
        "category": "",
        "price": "",
        "socialPrice": "",
    }

    frame = priceTag[math.floor((len(priceTag) / 2))]

    results = model.track(frame, iou=0.4, conf=0.5, persist=True, imgsz=608, verbose=False,
                          tracker="models/botsort.yaml")
    title = ""
    category = ""
    price = ""
    socialPrice = ""
    flag = False

    if results[0].boxes.id != None:
        boxes = results[0].boxes.xyxy.cpu().numpy().astype(int)
        ids = results[0].boxes.id.cpu().numpy().astype(int)

        for box, idx in zip(boxes, ids):
            cropped = frame[box[1]:box[3], box[0]:box[2]]

            if idx == 2:
                title += "".join(reader.readtext(cropped, detail=0, paragraph=True))
            elif idx == 1:
                flag_price = True
                readerr = reader.readtext(cropped, detail=0)
                for item in readerr:
                    if convert_to_int(item) != None and flag_price:
                        price = f'{convert_to_int(item)}'
                        flag_price = False
                    elif convert_to_int(item) != None and (not (flag_price)):
                        price = price + "." + f'{convert_to_int(item)}'
            elif idx == 3:
                socialPrice += "".join(reader.readtext(cropped, detail=0, paragraph=True))

    new_category = [preprocess_text(title)]  # List of new titles
    new_category_tfidf = ClassificationTitleTfidfVectorizer.transform(new_category)
    predictions = modelClassificationTitle.predict(new_category_tfidf)

    if (len(price.split(".")) < 3) and (len(price) != 0) and (len(price) != 1) and (float(price) < 1):
        dataPrice = price.split(".")
        price = dataPrice[1] + "." + dataPrice[0]

    if socialPrice is not None:
        flag == True

    return {
        "title": title,
        "category": predictions[0],
        "price": price,
        "socialPrice": flag,
    }


def process_score(file_paths):
    dfr = pd.DataFrame(columns=['Наименование файла', 'Категория продукта', 'Цена'])
    count = 3
    for file_path in os.listdir(file_paths):
        result = detectPriceNameSocialPrice(os.path.join(file_paths, file_path))
        dfr.loc[len(dfr.index)] = [file_path, result['category'], result['price']]  
        
        print("Completed", file_path)
        count -= 1
        if count == 0:
            break
        
    dfr.to_csv("output.csv", sep=';',encoding = 'utf-8', index = False)
            
process_score('train')
```
